```python
pb_range = np.arange(0.0025,.1,0.0025)

exp_dir = "exp_data/Freq1_Policy1/PB{}_SB0_HB0"

results_arr = []
for pb in pb_range:
    pb = round(pb,4)
    # print(pb)
    exp = exp_dir.format(pb)
    results_ = pd.read_csv("{}/results.csv".format(exp))
    results_["pb"] = pb
    results_arr.append(results_)

```

```python
len(results_arr[0])
```

```python
results = pd.concat(results_arr).reset_index(drop=True)
title = "Credential Linking: Periodic PCM, Positional noise"
sns.lineplot(results, x="pb", y="recall").set_title(title)
```

```python
title = "Credential Linking: Periodic PCM, Positional noise"
sns.lineplot(results, x="pb", y="prec").set_title(title)
```
